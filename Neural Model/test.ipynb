## Importing Required Library

```python
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
import joblib
```

```python
# defining model structure
class PredictRisk(nn.Module):
  def __init__(self, input_dim):
    super(PredictRisk, self).__init__()
    self.net = nn.Sequential(
        nn.Linear(input_dim,16),
        nn.ReLU(),
        nn.Dropout(p=0.4),

        nn.Linear(16,8),
        nn.ReLU(),
        nn.Dropout(p=0.4),

        nn.Linear(8,4),
        nn.ReLU(),
        nn.Dropout(p=0.4),

        nn.Linear(4,1),
        nn.Sigmoid()
    )

  def forward(self, x):
    return self.net(x)
```

```python
import warnings
from sklearn.exceptions import InconsistentVersionWarning
warnings.filterwarnings("ignore", category=InconsistentVersionWarning)
model_path = "/home/ritu/safetronics/Neural Model/NeuralModel.pth"
scaler_path = "/home/ritu/safetronics/Neural Model/Neural_Scaler.joblib"
scaler = joblib.load(scaler_path)
```

```python
input_dim = 9
model = PredictRisk(input_dim = input_dim)
```

```python
# load the state dict
state = torch.load(model_path)
model.load_state_dict(state)
model.eval()
```
